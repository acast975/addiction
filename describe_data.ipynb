```python
import matplotlib.pyplot as plot
import seaborn as sns
import helpers
```

```python
feature_data = helpers.load_clean_data('./datasets/addict-fix-limit.csv')
feature_data_count = len(feature_data)
feature_data.head()
```

```python
sns.set_theme(style="darkgrid")
for feature in feature_data:
    plot.figure(figsize=(10,7)) 
    ax = sns.countplot(x=feature, data=feature_data) 
    plot.title('Distribution of {}'.format(feature)) 
    plot.xlabel('Value') 
    plot.ylabel('Count (Frequency [%])') 
    for p in ax.patches: 
        ax.annotate('{0} ( {1:.2f}% )'.format(p.get_height(), 100 * p.get_height()/feature_data_count), (p.get_x()+0.2, p.get_height()+15))
    plot.show()
```

```python
sns.set_theme(style="darkgrid")
for feature in feature_data:
    if feature == 'PUIcutoff':
        continue
    plot.figure(figsize=(10,7)) 
    ax = sns.countplot(x=feature, data=feature_data, hue='PUIcutoff') 
    plot.title('Distribution of {}'.format(feature)) 
    plot.xlabel('Value') 
    plot.ylabel('Count (Frequency [%])') 
    for p in ax.patches: 
        ax.annotate('{0} ( {1:.2f}% )'.format(p.get_height(), 100 * p.get_height()/feature_data_count), (p.get_x()+0.2, p.get_height()+15))
    plot.show()
```
